```python
import pandas as pd

df = pd.read_csv("Estatisticas_Jogo.csv")

display(df)
```

```python
# Média de Pontos dos Jogadores

media_pontos = df["Pontos"].mean()

print(f"A média de pontos foi de {round(media_pontos, 1)} pontos")
```

```python
# Jogador com mais pontos

mais_pontos = df["Pontos"].idxmax()

print(f"O jogador com mais pontos foi o {df.loc[mais_pontos, "Jogador"]} com {df.loc[mais_pontos, "Pontos"]} pontos")
```

```python
#Copiar o ficheiro estatisticas_jogo e colar/guardar dentro do ficheiro ranking_jogadores
df.to_csv("Ranking_Jogadores.csv", index=False)

df_ranking = pd.read_csv("Ranking_Jogadores.csv")

df_ranking.sort_values(by="Pontos", ascending=True, inplace=True)

df_ranking.to_csv("Ranking_Jogadores.csv", index=False)

display(df_ranking)
```
