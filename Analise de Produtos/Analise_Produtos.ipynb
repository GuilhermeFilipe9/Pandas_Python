```python
import pandas as pd

df = pd.read_csv("Produtos.csv")

display(df)
```

```python
# Calcular valor Total de cada produto

df["Valor Total"] = df["Preço"] * df["Quantidade"]

df.to_csv("Produtos.csv", index=False)
```

```python
# Encontar o produto mais caro

produto_maior = df["Preço"].idxmax()

display(f"O Produto mais caro é o {df.loc[produto_maior, "Produto"]} com o valor de {df.loc[produto_maior, "Preço"]} €")
```

```python
# Ordenar Produtos por preço crescente

df.sort_values(by="Preço", ascending=True, inplace=True)

df.to_csv("Produtos.csv", index=False)

display(df)
```
